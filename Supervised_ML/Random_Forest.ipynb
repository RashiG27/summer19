```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
```

```python
df=pd.read_csv('social.csv')
df
```

```python
#picking features
features=df.iloc[:,:-1].values    #it will change df into numpy array
features
```

```python
label=df.iloc[:,-1].values
label
```

```python
actual_features=features[:,2:4]
actual_features.shape
```

```python
#feature scaling 
from sklearn.preprocessing import StandardScaler
f_sc=StandardScaler()
```

```python
#now training and training data sep
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(actual_features,label,test_size=0.2)
```

```python
# here apply feature scaling 
final_train=f_sc.fit_transform(train_x) #takes maximum of training data
```

```python
final_train
```

```python
#testing data 
final_test=f_sc.transform(test_x)   # only testing data
```

```python
final_test
```

```python
#calling RandomForest Classifier
clf=RandomForestClassifier(n_estimators=30, criterion='entropy')
```

```python
from sklearn.metrics import accuracy_score
```

```python
trained=clf.fit(train_x,train_y)
```

```python
predicted=trained.predict(test_x)
predicted
```

```python
test_y
```

```python
accuracy_score(predicted,test_y)
```
